```python
import psycopg2
from keys.db_keys import db_keys

conn = psycopg2.connect(**db_keys)
cur = conn.cursor()


def get_random_words():
    cur.execute("SELECT * FROM words ORDER BY RANDOM() LIMIT 1000;")
    return [i[1] for i in cur.fetchall()]

get_random_words()
```
